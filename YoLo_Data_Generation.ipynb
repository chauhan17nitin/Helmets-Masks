```
from xml.etree import ElementTree
```

```
tree = ElementTree.parse('annotations.xml')
root = tree.getroot()
```

```
for image in root.findall('.//image'):
  image_width = float(image.attrib['width'])
  image_height = float(image.attrib['height'])
  print(image_width, image_height)
  file1 = open('annotations/' + image.attrib['id' ] + '.txt',"a+")
  
  for box in image.findall('.//box'):
    if box.attrib['label'] == 'head':
      xtl = float(box.attrib['xtl'])
      ytl = float(box.attrib['ytl'])
      xbr = float(box.attrib['xbr'])
      ybr = float(box.attrib['ybr'])

      absolute_width = xbr - xtl
      absolute_height = ybr - ytl
      if xbr - xtl > 10 and ybr - ytl > 10:
        x_centre = xtl + (absolute_width)/2
        y_centre = ytl + (absolute_height)/2
        list1 = ['0']
        list1.append(str(round(x_centre/image_width, 6)))
        list1.append(str(round(y_centre/image_height, 6)))
        list1.append(str(round(absolute_width/image_width, 6)))
        list1.append(str(round(absolute_height/image_height, 6))+ '\n')
        file1.write(" ".join(list1))
  file1.close()
  # break
```
